```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("cardio_train.csv", sep=";")
df.head()
```

Construa um subset desse dataset, usando 1000 amostras da classe 0 (doença ausente) e 1000 amostras da classe 1 (doença presente).  Use 80% para o dataset de treino e 20% para o dataset de teste.

```python
# Confesso que eu ainda não entendi o motivo de separar 1000 amostras de cada. 
# Custo computacional? PS: Eu preciso juntar depois, saquei.
df_cls0 = df[df['cardio'] == 0].sample(n=1000, random_state=42)
df_cls1 = df[df['cardio'] == 1].sample(n=1000, random_state=42)
```

Nota: entendo que X (maiusculo) são minhas features. O que tenho/caracteristicas/dados de entrada. y é o rotulo e variavel alvo. O que eu quero prever.

X_train e y_train: dados de treino. É os dados pra aprendizado + gabarito. Ele vai olhar pros dados de treino X_train e pras respostas y_train.

X_test são os dados de teste. Depois que o modelo treino, ele recebe o X_test que sao apenas as perguntas sem respostas e com isso ele preve em cima. O y_test seria o gabarito que fica escondido. Quando o modelo devolver as previsoes feitas em X_test, eu comparo essas previsoes com o y_test verdadeiro. Essa etapa é o calculo da acuracia ou erro.

```python
df_sub = pd.concat([df_cls0, df_cls1])

X = df_sub.drop(columns=['cardio', 'id']) # tudo menos o que quero prever e os ids(ruido)
y = df_sub['cardio'] # o que quero prever

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=42,
)
```

```python
print(X_train.values)
print(y_train)
```

```python
for k in range(2, 10):
    knn = KNeighborsClassifier(n_neighbors=k)
    #knn.fit(X_train, y_train)
    knn.fit(X_train.values, y_train.values) # se eu uso o .values eu apro de receber warning
    # aparentemente eh mais  facil treinar o modelo usando numeros puros sem strs
    print(f"K={k}; Acc={knn.score(X_test.values, y_test)}")

    correctness = 0
    for i in range(0, len(X_test)): # enbtendo por X_test os novos "data points"
       # print("x_test.values=", [X_test.values[i]])
        predict = knn.predict([X_test.values[i]])
        #print(X_test.values[i], "prediction: ", predict)
        if predict == y_test.iloc[i]: # tive um problema de indexação entao troquei p .iloc[i]
            correctness += 1
    acc = correctness / len(X_test)
    print("acertos=", correctness)
    print("erros=", len(X_test) - correctness)
    print("accuracy=", acc)
```

Aparentemente o melhor resultado foi com k=9 com 59% de acc com maior numero absoluto de acertos.
