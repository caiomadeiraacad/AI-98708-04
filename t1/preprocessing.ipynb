```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import csv
```

Pra nao esquecer:
- [0, 1, 2], [3, 4, 5], [6, 7, 8], OPÇÕES DE VITORIAIS PELAS LINHAS
- [0, 3, 6], [1, 4, 7], [2, 5, 8], OPÇÕES DE VITORIAIS PELAS COLUNAS
- [0, 4, 8], [2, 4, 6]             OPÇÕES DE VITORIAIS PELAS DIAGONAIS

```python
win_scenarios = [
    [0, 1, 2], [3, 4, 5], [6, 7, 8],
    [0, 3, 6], [1, 4, 7], [2, 5, 8], 
    [0, 4, 8], [2, 4, 6]             
]
```

```python
def check_winner(board, player):
    for pos1, pos2, pos3 in win_scenarios:
        if board[pos1] == player and board[pos2] == player and board[pos3] == player:
            return True
    return False
```

```python
with open('data/tic-tac-toe.data', 'r') as f:
    lines = f.readlines()

    x_win = []
    o_win = []
    draws = []
    continue_game = []

    for line in lines:
        data = line.strip().split(',')
        board = data[0:9]
        original_class = data[9]

        if original_class == 'positive':
            data[9] = 'X_WIN'
            x_win.append(data)
        else:
            if check_winner(board, 'o'):
                data[9] = 'O_WIN'
                o_win.append(data)
            else:
                data[9] = 'Draw'
                draws.append(data)

    all_results = x_win + o_win + draws
    
    while len(continue_game) < 300:
        rdn = random.choice(all_results).copy()
        board = rdn[0:9]
        
        qtd_rm = random.randint(2, 5)
        
        occupied_indexes = [i for i in range(9) if board[i] != 'b']
        
        if len(occupied_indexes) < qtd_rm:
            continue
            
        rm = random.sample(occupied_indexes, qtd_rm)
        for i in rm:
            board[i] = 'b'
            
        qtd_x = board.count('x')
        qtd_o = board.count('o')
        
        if qtd_x == qtd_o or qtd_x == qtd_o + 1:
            if not check_winner(board, 'x') and not check_winner(board, 'o'):
                rdn[9] = 'Continue_Game'
                continue_game.append(rdn)

    df_final = []
    
    random.shuffle(x_win) # lembrar q shuffle embaralha
    random.shuffle(o_win)
    random.shuffle(draws)
    random.shuffle(continue_game)
    
    tr = 200
    df_final.extend(x_win[:tr])
    df_final.extend(o_win[:tr])
    df_final.extend(draws[:tr])
    df_final.extend(continue_game[:tr])
    
    random.shuffle(df_final)

    header = [
        'top_left_square', 'top_middle_square', 'top_right_square',
        'middle_left_square', 'middle_middle_square', 'middle_right_square',
        'bottom_left_square', 'bottom_middle_square', 'bottom_right_square',
        'class'
    ]
    
    with open('data/dataset_t1.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(df_final)
```
