```python
import matplotlib.pyplot as plt 
from sklearn.neighbors import KNeighborsClassifier
```

```python
x = [2, 4, 12, 8, 10, 12]
y = [6, 8, 5, 4, 23, 1]
classes = [1, 1, 0, 0, 0, 1]

plt.scatter(x, y, c=classes, cmap='plasma')
plt.show()
```

```python
data = list(zip(x, y)) # zip relaciona as componentes de um ponto; x com y;
print("data:", data) 
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(data, classes)
```

```python
# agora tentando classificar um novo ponto de dados
new_x = 6
new_y = 3
new_point = [(new_x, new_y)]

pred = knn.predict(new_point)

plt.scatter(x + [new_x], y + [new_y], c=classes + [pred[0]])
plt.text(x=new_x-1.7, y=new_y-0.7, s=f"new point, class: {pred[0]}")
plt.show()
```

tentando com um k_neighbors maior

```python
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(data, classes)
pred = knn.predict(new_point)

plt.scatter(x + [new_x], y + [new_y], c=classes + [pred[0]])
plt.text(x=new_x-1.7, y=new_y-0.7, s=f"new point, class: {pred[0]}")
plt.show()
```
